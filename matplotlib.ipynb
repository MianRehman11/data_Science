```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 2D array (e.g., a grid of values)
data = np.array([[5, 6, 7], 
                 [8, 9, 2], 
                 [3, 1, 4]])

# Display the data using imshow
plt.imshow(data, cmap='hot', interpolation='nearest')
plt.colorbar()  # Show a color bar
plt.title('Heatmap Example')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate some random data
x = np.random.rand(10)
y = np.random.rand(10)
z = np.random.rand(10)

# Create a new figure
fig = plt.figure()

# Add a 3D subplot
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D scatter plot
ax.scatter(x, y, z, c='r', marker='o')  # 'c' for color and 'marker' for the style

# Add labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.title('3D Scatter Plot Example')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data
data = [np.random.normal(loc=0, scale=1, size=100),  # Group 1
        np.random.normal(loc=1, scale=1.5, size=100),  # Group 2
        np.random.normal(loc=2, scale=0.5, size=100)]  # Group 3

# Create a violin plot
plt.violinplot(data)

# Add labels and title
plt.title('Violin Plot Example')
plt.xlabel('Groups')
plt.ylabel('Values')

# Show the plot
plt.xticks([1, 2, 3], ['Group 1', 'Group 2', 'Group 3'])  # Custom x-tick labels
plt.show()
```
