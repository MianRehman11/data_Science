```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data = {
    'Age' : [45,32,50,29,55,40,60,37,49,30,12,45,67,87,93,12,34,56,78,65],
    'BMI' : [28.3,32.1,32.1,24.2,36.2,21.3,24.1,21.6,31.9,24,23.6,12.5,32.3,25.2,32.1,29.1,19.6,32.9,21.2,12.8],
    'Blood pressure' : [130,145,123,124,125,184,123,90,130,140,123,140,213,123,132,143,173,153,210,212],
    'Cholesterol level' : [140,213,123,132,143,173,153,210,212,130,145,123,124,125,184,123,190,130,210,199],
    'Physical activity level' : [2.5,1,3,5,67,8,9,2,3,4,2.8,1,2,3,4,9,5,1,8,4],
    'Family history of the disease' : ['yes','yes','yes','yes','yes','yes','yes','yes','yes','yes','no','no','no','no','no','no','no','no','no','no']
}
```

```python
X = data.iloc[:,:-1]
```

```python
y = data.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classfier = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'penalty':['l2','l1','elasticnet'],'C':[1,2,3,4,5,6],'max_iter':[100,200,300,400,500]}          
```

```python
data1 = GridSearchCV(classfier,parameters,scoring='accuracy',cv=5)
```

```python
data1.fit(X_train,y_train)
```

```python
data1.best_estimator_
```

```python
data1.best_index_
```

```python
data1.best_params_
```

```python
data1.best_score_
```

```python
y_pri = data1.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
clr_report = classification_report(y_pri,y_test)
```

```python
clr_report
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_pri,y_test)
```

```python
cm
```

```python
sns.heatmap(cm,annot=True)
plt.xlabel('pridiction')
plt.ylabel('Truth')
plt.title('Heatmap')
plt.show()
```
