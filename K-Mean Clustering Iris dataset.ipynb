```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("Iris.csv")
```

```python
df.head(5)
```

```python
df = df.drop(["Id","Species"],axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
ss
```

```python
scale = ss.fit_transform(df)
```

```python
scale
```

```python
sns.pairplot(df)
```

```python
plt.scatter(df["SepalLengthCm"],df["SepalWidthCm"])
plt.xlabel("SepalLength (Cm)")
plt.ylabel("SepalWidth (cm)")
plt.show()
```

```python
km = KMeans(n_clusters=3)
predict = km.fit_predict(scale)
```

```python
predict
```

```python
df["Cluster"] = predict
```

```python
df
```

```python
km.cluster_centers_
```

```python
sns.pairplot(df,vars = ["SepalLengthCm",'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'],hue = "Cluster" ,palette="viridis")
plt.suptitle("Pair Plot with Cluster - Iris Dataset" , y=1.07)
plt.show()
```

```python
centriods = km.cluster_centers_
plt.scatter(scale[:,0],scale[:,1],c = df["Cluster"] , cmap="viridis",marker="o" , label = "Data Points")#ager kese data ko different color ma krna ho to c ko os column ka baraber rakhta ha
plt.scatter(centriods[:,0],centriods[:,1],c = 'red',marker="o" , label = "Centroids")
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.title("K-Mean Clustring with Centroids")
plt.legend()
plt.show()
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k,random_state=42)
    km.fit(df[['SepalLengthCm','SepalWidthCm']])
    sse.append(km.inertia_)

plt.xlabel("Number of Cluster (K)")
plt.ylabel("Sum of squared errors (SSE)")
plt.plot(k_rng,sse)
plt.show()
```
