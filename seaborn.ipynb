```python
arr = np.array([['a1','a2','a3','a4','a5'],
                 ['b1','b2','b3','b4','b5']])
```

```python
arr = np.array([['a1','a2','a3','a4','a5'],
                 ['b1','b2','b3','b4','b5']])
ar1 = np.linspace(1,10,10).reshape(2,5)
a = sns.heatmap(ar1,annot=arr,fmt = 's',linewidth = 5,linecolor = 'r')
```

### Styling plot

```python
var = sns.load_dataset('tips')
```

```python
var
```

```python
sns.set_style("darkgrid")
sns.barplot(x = "day",y = "total_bill",data = var)
plt.show()
```

```python
sns.set_style("white")
sns.barplot(x = "day",y = "total_bill",data = var)
sns.despine()
plt.show()
```

```python

sns.set_style("white")
sns.set_context('poster',font_scale= 0.3)
sns.barplot(x = "day",y = "total_bill",data = var,color = "r",saturation=0.2)
# sns.despine()
plt.show()
```

```python

sns.set_style("white")
sns.set_context('paper',font_scale= 1)
sns.barplot(x = "day",y = "total_bill",data = var,palette='Dark2')
# sns.despine()
plt.show()
```

### Multi plot

```python
var = sns.load_dataset('tips')
var
```

```python
fg = sns.FacetGrid(var,col = 'sex')
fg.map(plt.scatter,'total_bill','tip')
plt.show()
```

```python
fg = sns.FacetGrid(var,col = 'sex',hue = 'day')
fg.map(plt.bar,'total_bill','tip').add_legend()
plt.show()
```
